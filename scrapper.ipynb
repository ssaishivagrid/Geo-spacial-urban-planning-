```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

url = "https://en.wikipedia.org/wiki/Provinces_of_Mongolia"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find("table", {"class": "wikitable"})

capital_cities = []

for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    if len(columns) > 1:
        
        capital = columns[5].text.strip()
        
        capital = capital.split('(')[0].strip()
        
        capital_cities.append(f"{capital}, Mangolia")

df = pd.DataFrame(capital_cities, columns=["City, Country"])

csv_filename = "indian_state_capitals.csv"

file_exists = os.path.isfile(csv_filename)

df.to_csv(csv_filename, mode='a', header=not file_exists, index=False)

print(f"Data appended to CSV file: {csv_filename}")
```
